```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('gdrive/My Drive/APM/P2/Car_sales.csv')
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
rata_hilang = df['__year_resale_value'].mean()
df['__year_resale_value'] = df['__year_resale_value'].fillna(rata_hilang)
df['__year_resale_value'].isna().sum()
```

```python
rata_hilang = df['Price_in_thousands'].mean()
df['Price_in_thousands'] = df['Price_in_thousands'].fillna(rata_hilang)
df['Price_in_thousands'].isna().sum()
```

```python
rata_hilang = df['Engine_size'].mean()
df['Engine_size'] = df['Engine_size'].fillna(rata_hilang)
df['Engine_size'].isna().sum()
```

```python
rata_hilang = df['Horsepower'].mean()
df['Horsepower'] = df['Horsepower'].fillna(rata_hilang)
df['Horsepower'].isna().sum()
```

```python
rata_hilang = df['Wheelbase'].mean()
df['Wheelbase'] = df['Wheelbase'].fillna(rata_hilang)
df['Wheelbase'].isna().sum()
```

```python
rata_hilang = df['Width'].mean()
df['Width'] = df['Width'].fillna(rata_hilang)
df['Width'].isna().sum()
```

```python
rata_hilang = df['Length'].mean()
df['Length'] = df['Length'].fillna(rata_hilang)
df['Length'].isna().sum()
```

```python
rata_hilang = df['Curb_weight'].mean()
df['Curb_weight'] = df['Curb_weight'].fillna(rata_hilang)
df['Curb_weight'].isna().sum()
```

```python
rata_hilang = df['Fuel_capacity'].mean()
df['Fuel_capacity'] = df['Fuel_capacity'].fillna(rata_hilang)
df['Fuel_capacity'].isna().sum()
```

```python
rata_hilang = df['Fuel_efficiency'].mean()
df['Fuel_efficiency'] = df['Fuel_efficiency'].fillna(rata_hilang)
df['Fuel_efficiency'].isna().sum()
```

```python
rata_hilang = df['Power_perf_factor'].mean()
df['Power_perf_factor'] = df['Power_perf_factor'].fillna(rata_hilang)
df['Power_perf_factor'].isna().sum()
```

```python
df.isnull().sum()
```

```python
df.describe().T
```

```python
df["Total_power"]=df["Horsepower"]+df["Fuel_capacity"]+df["Fuel_efficiency"]
df=df[df["Total_power"]>0]
```

```python
agg_df = df.groupby(['Manufacturer', 'Price_in_thousands'])[["Total_power"]].aggregate("mean")
agg_df= agg_df.sort_values('Total_power', ascending=False)
agg_df = agg_df.reset_index()
agg_df.head()
```

```python
df["Segment"]=pd.qcut(df['Total_power'], 4, labels=['D','C','B','A'])
df[["Segment","Total_power"]].groupby("Segment").agg(["mean","min","max","sum"])
```

```python
agg_df =pd.DataFrame(df.groupby(['Manufacturer','Model'])['Total_power'].agg('mean')).round(2).sort_values("Total_power")
agg_df.head(10)
```

```python
agg_df=agg_df.reset_index()
agg_df.head()
```

```python
agg_df['customers_level_based'] = [str(row[0]).upper()+'_'+str(row[1]).upper() for row in agg_df.values]
agg_df = agg_df[["customers_level_based","Total_power"]].groupby("customers_level_based").agg("mean").round(2)
agg_df=agg_df.reset_index()
agg_df.head()
```

```python
agg_df["SEGMENT"] = pd.qcut(agg_df['Total_power'],4,labels=['D','C','B','A'])
agg_df
```
